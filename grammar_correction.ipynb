```python
from happytransformer import HappyTextToText
```

```python
happy_tt = HappyTextToText("T5", "prithivida/grammar_error_correcter_v1")
```

```python
from happytransformer import TTSettings
```

```python
top_k_sampling_settings = TTSettings(do_sample=True, top_k=50, temperature=0.7, min_length=1, max_length=50)
```

```python
result_1 = happy_tt.generate_text("They likes to eat apples", args=top_k_sampling_settings)
```

```python
print(result_1.text)
```

```python
result_2 = happy_tt.generate_text("I wentt too the stores to buys food.", args=top_k_sampling_settings)
```

```python
print(result_2.text)
```
